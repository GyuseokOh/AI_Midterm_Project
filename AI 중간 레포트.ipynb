```python
import numpy as np
from google.colab import drive
drive.mount('/content/drive')

train_data_path="/content/drive/MyDrive/인공지능 중간 레포트/train_data.npy"
train_label_path="/content/drive/MyDrive/인공지능 중간 레포트/train_label.npy"

train_data=np.load(train_data_path,allow_pickle=True)
train_data_rawdata=np.load(train_label_path,allow_pickle=True)

num_train_data=len(train_data)
```

```python
max_len=-1
max_index=None
min_len=10000000
min_index=None

for index, lists in enumerate(train_data):
  list_len=len(lists)
  if(list_len>max_len):
    max_index=index
    max_len=list_len
  if(list_len<min_len):
    min_index=index
    min_len=list_len

print("shortest index : %d, shortest length : %d"%(min_index,min_len))
print("longest index: %d, longest length : %d"%(max_index,max_len))
```

```python
import random

count=0;
check=[]

for each_game in train_data[random.randrange(len(train_data))]:
  if(count>=10):
    break
  each_game.resize(19,19)
  for row_index, row_list in enumerate(each_game):
    if(count>=10):
      break
    for col_index, col_element in enumerate(row_list):
      if(count>=10):
        break
      cur_point=[row_index,col_index]
      if(col_element!=0 and cur_point not in check):
        print("%dst : [%d, %d]"%(count+1, row_index,col_index))
        check.append(cur_point)
        count+=1
```

```python
max_b=-1
max_all=0
max_index=0

for game_index, game in enumerate(train_data):
  datalen=len(game)
  last_game=game[datalen-1]
  last_game.resize(19,19)
  all_count=0
  b_count=0

  for row_index, row in enumerate(last_game):
    for col_index, element in enumerate(row):
      if(element!=0):
        if(row_index<9 and col_index<9):
          b_count+=1
        all_count+=1

  if(b_count>max_b):
    max_b=b_count
    max_all=all_count
    max_index=game_index

print("max B index :%d"%(max_index))
print("max B : %d, A : %d, B/A : %.4f"%(max_b,max_all,max_b/max_all))
```
